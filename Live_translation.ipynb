# **Realtime translation**

This is a very basic Live tranalation application. It produces texual translations. It is uses a very basic pipeline:

    Audio stream --> ASR model --> MT model

 **NOTE:** 
 * The colab's T4 is a very basic GPU for running bigger models for realtime appliction and would introduce significant latency. Hence, we use  the wav2vec2-CTC model. If you have access to better GPU, we recommend using Whisper-large-v3-turbo*


* Additionally We use IndicTrans2-indic-en-dist-200M from Ai4bharat, for tranlation. You can upgrade to the 1B model that they offer, or use the NLLB if you want non indian languages too.`

### Prerequisites:

1. You must include your huggingface token using - `secret tab` (the key button on the left) to colab.

2. IndicTrans2 is a gated model so you must go request access - [here](https://huggingface.co/ai4bharat/indictrans2-en-indic-dist-200M)!

3. Change the runtype to GPU -T4
